```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
newdf = df.drop(columns = ['Name','Ticket','Cabin','PassengerId','Pclass','SibSp','Parch'])
df1 = pd.get_dummies(newdf,dtype=int)
```

```python
df1.corr()
```

```python
x1 = df1.drop(columns = ['Survived'])
y1 = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size=0.33)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
n = int(input())
for i in range(n):
    Sex_female,Sex_male,Embarked_C,Embarked_Q,Embarked_S = map(int,input().split(','))
    Age,Fare = map(float,input().split(','))
    print(model.predict([[Age,Fare,Sex_female,Sex_male,Embarked_C,Embarked_Q,Embarked_S]]))
```
